```python
import pandas as pd
import numpy as np
data_input = pd.ExcelFile('Hasil Generate Unique 2.xlsx').parse(0)
X_data = data_input.Vector2.to_numpy()
y = data_input.WazanIndex.to_numpy()
X = []
for i in range(0,len(X_data)):
    X.append(np.fromstring(X_data[i],sep=','))
X = np.asarray(X,dtype=object)
X.shape
```

```python
X = pd.DataFrame(X, columns=['P', 'I1','A1', 'I2','A2', 'I3','A3', 'I4','A4', 'I5','A5','S'])
y = pd.Series(y)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.2)
print(list(set(y_train)))
print(len(list(set(y_train))))
print(list(set(y_test)))
print(len(list(set(y_test))))
```
